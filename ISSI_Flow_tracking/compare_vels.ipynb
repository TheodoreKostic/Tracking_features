```python
import numpy as np 
import matplotlib.pyplot as plt 
from astropy.io import fits 
from scipy.stats import pearsonr
```

```python
# Referent Muram velocity:

def mean_v_from_file(file):
    v_x = fits.open(file)[0].data
    v_x.shape
    v_x = np.mean(v_x, axis=0)
    small = np.where(v_x < -10.0)
    v_x[small] = -10.0
    big = np.where(v_x > 10.0)
    v_x[big] = 10.0
    return v_x
```

```python
v_ic_x = mean_v_from_file("ic_fwhm_600_dt_30_tracked.fits")
v_ic_x.shape
```

```python
v_lw_default_x = mean_v_from_file("lw_500nm_fwhm_600_dt_30_tracked.fits")
v_lw_default_x.shape
```

```python

plt.figure(figsize=[17,5])
plt.subplot(131)
plt.imshow(v_ic_x.T, origin='lower', cmap='bwr', vmin=-5,vmax=5)
plt.subplot(132)
plt.imshow(v_lw_default_x.T, origin='lower', cmap='bwr', vmin=-5,vmax=5)
plt.subplot(133)
plt.scatter(v_ic_x.flatten(), v_lw_default_x.flatten(), alpha=0.02)
plt.xlim([-5,5])
plt.ylim([-5,5])

print(i, pearsonr(v_ic_x.flatten(), v_lw_default_x.flatten()))
```
