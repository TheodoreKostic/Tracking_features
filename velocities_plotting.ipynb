```python
import matplotlib.pyplot as plt 
import numpy as np 
from astropy.io import fits 
```

```python
vx = fits.open("velocities.fits")[0].data
vy = fits.open("velocities.fits")[1].data
vxt = fits.open("velocities.fits")[2].data
vyt = fits.open("velocities.fits")[3].data
vxb = fits.open("velocities.fits")[4].data
vyb = fits.open("velocities.fits")[5].data
```

```python
vx.shape
```

```python
# Check the completely averaged values between the two: 

vxmean = np.mean(vx,axis=0)
vymean = np.mean(vy,axis=0)
vxtmean = np.mean(vxt,axis=0)
vytmean = np.mean(vyt,axis=0)
vxbmean = np.mean(vxb,axis=0)
vybmean = np.mean(vyb,axis=0)
```

```python
plt.figure(figsize=[12,5])
plt.subplot(121)
plt.imshow(vxmean.T,origin='lower',cmap='bwr',vmin=-5,vmax=5)
plt.subplot(122)
plt.imshow(vxtmean.T,origin='lower',cmap='bwr',vmin=-2,vmax=2)
```

```python
plt.scatter(vxmean.T.flatten(), vytmean.T.flatten()*3, alpha=0.005)
plt.xlim([-3,3])
plt.ylim([-3,3])
```
